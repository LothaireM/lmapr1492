```python
from pymatgen.ext.matproj import MPRester
from pymatgen.analysis.diffraction.xrd import XRDCalculator
import matplotlib.pyplot as plot
```

```python
with MPRester("D0ZXquWb8kOQRgLT") as m:
    # Structure for material id
    structure = m.get_structure_by_material_id("mp-1008559")
```

```python
xrdCalc = XRDCalculator('CuKa')
pattern = xrdCalc.get_pattern(structure)
Angle = pattern.x
Intensité = pattern.y
hkl = pattern.hkls
```

```python
#Premier pic du diffractogramme
print("\n Premier pic du diffractogramme")
print("Angle 2theta =",Angle[0],"°")
print("Intensité =",Intensité[0],"(cps)")
print("Indices hkl associés =",hkl[0][0]['hkl'])

#Deuxième pic du diffractogramme
print("\n Deuxième pic du diffractogramme")
print("Angle 2theta =",Angle[1],"°")
print("Intensité =",Intensité[1],"(cps)")
print("Indices hkl associés =",hkl[1][0]['hkl'])

#Troisième pic du diffractogramme
print("\n Troisième pic du diffractogramme")
print("Angle 2theta =",Angle[2],"°")
print("Intensité =",Intensité[2],"(cps)")
print("Indices hkl associés =",hkl[2][0]['hkl'])

#Plot du graphe des trois pics du diffractogramme
print("\n Plot du graphe des trois pics du diffractogramme :")
xrdCalc.show_plot(structure, two_theta_range=(20,40))
```
